```python
import pandas as pd
import numpy as np #numerical operation
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Import dataset


```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv', encoding = 'latin1')
```

```python
statistic = data.describe()
data.head()

x = "Previous_Exam_Scores"
y = "Grades"
print(data.head())
```

```python
plt.scatter(data[x], data[y])
plt.xlabel('Previous_Exam_Scores')
plt.ylabel('Grades')
plt.title('Test')
plt.show
```

```python
x = data [[x]]
y = data[y]
```

```python
model = LinearRegression()
```

```python
slope = model.coef_[0]
intercept = model.intercept_
print('slope', slope)
print('intercept',intercept)

plt.scatter(x,y)
plt.plot(x,model.predict(x), color = 'red')
plt.xlabel('Previous_Exam_Score')
plt.ylabel('Grades')
plt.title('Test')
plt.show()
```

```python
x = 'Attendance_Percentage'
y = 'Grades'

print(data.head())
x = data [[x]]
y = data [[y]]

model = LinearRegression()
model.fit(x,y)

```

```python
slope = model.coef_[0]
intercept = model.intercept_
print("slope", slope)
print("intercept", intercept)

plt.scatter(x,y)
plt.plot (x,model.predict(x),color = 'black')
plt.xlabel ('Independent Variable')
plt.ylabel ('Dependent Variable')
plt.title ('Hours of Study')
plt.show()
```
